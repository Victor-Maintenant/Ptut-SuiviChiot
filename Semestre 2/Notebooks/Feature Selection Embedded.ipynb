### Cible 'Died'

#### Ridge

```python
data = pd.read_csv("DataM.csv")
names=['litt_size','TPI','Diarr','hospit','InfCroissPed', 'InfCroiss4', 'InfCroiss5', 'InfCroiss6', 'InfCroiss7', 'InfCroiss8', 'TRD21', 'TRD22', 'TRD23', 'TRD24', 'TRD25', 'TRD26', 'TRD27', 'TRD28', 'TRD29', 'TRD30', 'TRD31', 'TRD32', 'TRD33', 'TRD34', 'TRD35', 'TRD36', 'TRD37', 'TRD38', 'TRD39', 'TRD40', 'TRD41', 'TRD42', 'TRD43', 'TRD21', 'TRD44', 'TRD45', 'TRD46', 'TRD47', 'TRD48', 'TRD49', 'TRD50', 'TRD51', 'TRD52', 'TRD53', 'TRD54', 'TRD55', 'TRD56']
data.fillna(0, inplace=True)
del data['ID']
del data['ID_litter']
del data['group']
del data['IgGJ2']
del data['AgeD2']
del data['Age_D1']
del data['Min1inflcr']
del data['Size']
del data['sex']
data.head()
```

```python
from sklearn.linear_model import Ridge
X=data.drop(["Died"], axis = 1)
Y=data['Died']
ridge = Ridge(alpha=1.0)
ridge.fit(X,Y)
```

```python
ridge=Ridge(alpha=1.0)
ridge.fit(X,Y)
predictors=X.columns
coef=pd.Series(ridge.coef_,predictors).sort_values(ascending=False)
print(coef)
```

#### Lasso

```python
from sklearn.linear_model import Lasso
lasso=Lasso(alpha=1)
lasso.fit(X,Y)
predictors=X.columns
coef=pd.Series(lasso.coef_,predictors).sort_values()
print(coef)
```
